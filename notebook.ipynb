## EVALUATION METRIC RESULTS

### Gather eval data in one dataframe

```python
import os
import pandas as pd

IGNOPRE_RUNS = ["minimal_example", "minimal_example_multi"]
eval_dfs = []
for path, dirs, files in os.walk('out'):
    for f in files:
        if f == 'eval_results.csv':
            # out/run/feature_extractor/split/..
            ps = path.split('/')
            run_name = ps[1]
            if run_name in IGNOPRE_RUNS:
                continue
            eval_df = pd.read_csv(f"{path}/{f}", index_col=[0])
            eval_dfs.append(eval_df)
evals = pd.concat(eval_dfs, axis=0, ignore_index=True)
display(evals)
evals.to_csv("out/all_metric_results.csv")
```
